# Exploring the Indian English + Hindi TTS Speech Dataset

555 curated, single-speaker clips (73.2 minutes total) with human-corrected transcripts, acoustic QC metrics, and speaking-style tags.

Full curation pipeline: https://github.com/auraCodesKM/sarvam-tts-dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import IPython.display as ipd

DATA_DIR = "/kaggle/input/indian-english-hindi-tts-dataset"
df = pd.read_csv(f"{DATA_DIR}/manifest.csv")
df.head()
```

## Language and duration

```python
print(df.groupby("language")["duration_s"].agg(["count", "sum"]).rename(columns={"sum": "total_seconds"}))

fig, axes = plt.subplots(1, 2, figsize=(11, 4))
df["language"].value_counts().plot.bar(ax=axes[0], title="clips by language")
df["duration_s"].plot.hist(bins=30, ax=axes[1], title="clip duration (s)")
plt.tight_layout()
plt.show()
```

## Acoustic quality (SNR) and speaking style

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
df["snr_db"].plot.hist(bins=30, ax=axes[0], title="SNR (dB)")
df["emotion"].value_counts().plot.bar(ax=axes[1], title="clips by speaking style")
plt.tight_layout()
plt.show()
```

## ASR quality: raw ASR vs. human-corrected transcript

```python
edited = df.groupby("language")["edited"].mean() * 100
print("% of clips where the human reviewer changed the ASR transcript:")
print(edited.round(1))
```

## Listen to a sample clip per language

```python
for lang in ["en-IN", "hi-IN"]:
    row = df[df.language == lang].iloc[0]
    print(lang, "-", row.text)
    display(ipd.Audio(f"{DATA_DIR}/{row.audio_path.split('data/')[-1]}"))
```
